# рџ”ґ Hard: Linear Self-Attention

Implement **Linear Attention** вЂ” O(SВ·DВІ) instead of O(SВІВ·D), enabling efficient long-sequence processing.

Replace softmax with a **kernel feature map** $\phi$:

$$\text{LinearAttn}(Q,K,V) = \frac{\phi(Q) \left(\phi(K)^T V\right)}{\phi(Q) \cdot \sum \phi(K)}$$

### Feature map
Use $\phi(x) = \text{elu}(x) + 1$ (ensures non-negative features).

### Key insight
Instead of computing the $S \times S$ attention matrix, compute $\phi(K)^T V$ first (a $D_k \times D_v$ matrix), then multiply by $\phi(Q)$.

### Rules
- Must use a feature map (NOT softmax)
- Must be O(SВ·DВІ) вЂ” should run fast on long sequences
- You **may** use `F.elu`

````python
import torch
import torch.nn.functional as F
````

````python
# вњЏпёЏ YOUR IMPLEMENTATION HERE

def linear_attention(Q, K, V):
  Q = F.elu(Q) + 1
  K = F.elu(K) + 1

  num = torch.bmm(Q, torch.bmm(K.transpose(-2, -1), V))
  den = torch.bmm(Q, (K.sum(dim=1, keepdim=True)).transpose(1, 2))
  output = num/den
  return output

    # pass  # Replace this
````
