````python
import torch
import torch.nn.functional as F
````

````python
from os import W_OK
# ✏️ YOUR IMPLEMENTATION HERE

def my_conv2d(x, weight, bias=None, stride=1, padding=0):
  x = F.pad(x, [padding]*4)
  N, C, H_in, W_in = x.shape
  Nk, Ck, Hk, Wk = weight.shape
  H_out = ((H_in - Hk)//stride) + 1
  W_out = ((W_in - Wk)//stride) + 1
  patches = x.unfold(2, Hk, stride).unfold(3, Wk, stride)
  out = torch.einsum('bihwjk,oijk->bohw', patches, weight)
  if bias is not None:
    out = out + bias.view(1, -1, 1, 1)
  return out

    # pass  # extract patches, apply kernel, handle stride/padding
````
